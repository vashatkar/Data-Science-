### Import datasets

```python
import json
import os

import numpy as np
import pandas as pd
from tqdm import tqdm

data = {'review_id' : [], 'business_id' : [], 'user_id' :[] ,'stars': [], 'text': []}

with open('C:/Users/vasha/OneDrive/unsupervised/Project/Data/review.json',encoding="utf8") as f:
    for line in tqdm(f):
        review = json.loads(line)
        data['review_id'].append(review['review_id'])
        data['business_id'].append(review['business_id'])
        data['user_id'].append(review['user_id'])
        data['stars'].append(review['stars'])
        data['text'].append(review['text'])
        
df = pd.DataFrame(data)

print(df.shape)
df.head()
df['review_id'] = df['review_id'].astype(str)
df['business_id'] = df['business_id'].astype(str)
df['user_id'] = df['user_id'].astype(str)
df['stars'] = df['stars'].astype(float)
df['text'] = df['text'].astype(str)
df.to_csv('C:/Users/vasha/OneDrive/unsupervised/Project/Data/yelp_reviews.csv', index=False)
```

```python
df.head()
```

```python


bussiness_data = {'state' : [] , 'name': [], 'business_id' : [], 'business_stars' : [] }


with open('C:/Users/vasha/OneDrive/unsupervised/Project/Data/business.json',encoding="utf8") as f:
    for line in tqdm(f):
        bussiness = json.loads(line)
        bussiness_data['state'].append(bussiness['state'])
        bussiness_data['name'].append(bussiness['name'])
        bussiness_data['business_id'].append(bussiness['business_id'])
        bussiness_data['business_stars'].append(bussiness['stars'])
        
bussiness_df = pd.DataFrame(bussiness_data)

print(bussiness_df.shape)
bussiness_df.head()

bussiness_df['name']

bussiness_df['state'] = bussiness_df['state'].astype(str)
bussiness_df['name'] = bussiness_df['name'].astype(str)
bussiness_df['business_id'] = bussiness_df['business_id'].astype(str)
bussiness_df['business_stars'] = bussiness_df['business_stars'].astype(float)

bussiness_df.to_csv('C:/Users/vasha/OneDrive/unsupervised/Project/Data/yelp_business.csv', index=False)
```
