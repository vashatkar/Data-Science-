```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
results_df = pd.read_csv(r'results_dataframe.csv', index_col = [0])
```

```python
results_df = results_df.reset_index().drop(['index'] , axis = 1)
```

```python
results_df
```

```python
results_df = results_df.drop([6, 0]).reset_index().drop(['index'], axis = 1 )
```

```python
results_df.iloc[5, 0] = 'Baseline'
```

```python
results_df
```

```python
results_df = results_df.sort_values(by='mean', ascending = False)
results_df
```

```python
results_final_df = pd.melt(results_df, id_vars=['Model_Name'], value_vars=['Fold 1', 'Fold 2', 'Fold 3','Fold 4', 'Fold 5', 'mean', 'std. dev'],
       var_name='type', value_name='values')

results_final_df

```

```python
# results_final_df.filter()
```

```python
results_final_df_plot1 = results_final_df.loc[ results_final_df['type'].isin(['Fold 1', 'Fold 2', 'Fold 3', 'Fold 4', 'Fold 5'])  ]

results_final_df_plot1.columns = ['Model_Name', 'Folds', 'values']

results_final_df_plot1
```

```python

plt.figure(figsize = (10, 10))
sns.barplot(x="Model_Name", y="values", hue="Folds", data=results_final_df_plot1)
plt.ylim([1, 2.5])
plt.ylabel("RMSE value")
plt.title("RMSE of 5 folds for all recommender system models")
plt.show()
```

```python
plt.figure(figsize = (20, 20)) 
sns.catplot(x="Model_Name", y="values",
                 hue="Folds",
                 data=results_final_df_plot1, kind="bar",
                 height=10, aspect=.7)
plt.show()
```

```python
results_final_df.loc[ results_final_df['type'].isin(['mean'])  ]
```

```python
results_final_df_plot2 = results_final_df.loc[ results_final_df['type'].isin(['mean'])  ]

results_final_df_plot2.columns = ['Model_Name', 'Mean', 'values']

results_final_df_plot2 = results_final_df_plot2.sort_values(by ='values', ascending = False)
results_final_df_plot2
```

```python
plt.figure(figsize = (10, 10))
barplot2 = sns.barplot(x="Model_Name", y="values", data=results_final_df_plot2)
# sns.lineplot(x="Model_Name", y="values",
#                   data= results_final_df_plot2)
barplot2.set_xticklabels(labels = results_final_df_plot2['Model_Name'],rotation = 45)
plt.xlabel('Model Name')
plt.ylabel('Average RMSE Value')
plt.title("Avearge RMSE for all recommender systems")
plt.show()
```

```python
plt.figure(figsize= (10, 10))
plt.plot(results_final_df_plot2['Model_Name'].values, results_final_df_plot2['values'].values, 
     color='GREEN', marker='o', 
     linestyle='dashed', linewidth=3, markersize=12)
for a,b in zip(results_final_df_plot2['Model_Name'].values , results_final_df_plot2['values'].values): 
    plt.text(a, b, str(b), fontsize=20)
plt.xlabel('Model Name')
plt.ylabel('Average RMSE Value')
plt.title("Avearge RMSE for all recommender systems")
plt.show()
```

```python
results_final_df_plot3 = results_final_df.loc[ results_final_df['type'].isin(['std. dev'])  ]

results_final_df_plot3.columns = ['Model_Name', 'std. dev', 'values']

results_final_df_plot3 = results_final_df_plot3.sort_values(by ='Model_Name')
results_final_df_plot3
```

```python
plt.figure(figsize = (10, 10))
barplot3 = sns.barplot(x="Model_Name", y="values", data=results_final_df_plot3)
# sns.lineplot(x="Model_Name", y="values",
#                   data= results_final_df_plot2)
barplot3.set_xticklabels(labels = results_final_df_plot2['Model_Name'],rotation = 45)
plt.xlabel('Model Name')
plt.ylabel('Standard deviation Value')
plt.title("Standard Deviation of RMSE for all recommender systems")
plt.show()
```

```python
plt.figure(figsize= (10, 10))
plt.plot(results_final_df_plot3['Model_Name'].values, results_final_df_plot3['values'].values, 
     color='red', marker='o', 
     linestyle='dashed', linewidth=3, markersize=12)
for a,b in zip(results_final_df_plot3['Model_Name'].values , results_final_df_plot3['values'].values): 
    plt.text(a, b, str(b), fontsize=20)
plt.xlabel('Model Name')
plt.ylabel('Standard deviation Value')
plt.title("Standard Deviation of RMSE for all recommender systems")
plt.show()
```

```python

results_final_df_plot4 = results_final_df_plot1.groupby('Model_Name').mean()


```

```python
results_final_df_plot4
```

```python
results_final_df_plot5 = results_final_df_plot1.merge(results_final_df_plot4, left_on='Model_Name', right_on='Model_Name',
           suffixes=('_left', '_right'))
```

```python
results_final_df_plot5
```

```python
results_final_df_plot5.loc[: , 'std_dev'] = results_final_df_plot5.loc[: , 'values_left'] - results_final_df_plot5.loc[: , 'values_right']
```

```python
results_final_df_plot5
```

```python
plt.figure(figsize = (10, 10))
sns.barplot(x="Model_Name", y="std_dev", hue="Folds", data=results_final_df_plot5)
plt.ylim([-0.010, 0.010])
plt.ylabel("RMSE standard deviation value")
plt.title("standard deviation of RMSE in 5 folds for all recommender system models (from mean of each model)")
plt.show()
```
