## Data preparation

```python

def preprocess_data(X ,way='keras'):
    if way == 'keras':
        # X_post = preprocess_input(X)
        X_post = X/255.
    else:
        X_post = X/255.
    return X_post
        
```

```python
np.random.seed(10)
def readData(root_directory):
    prefix = root_directory

    images = []
    labels = defaultdict(list)

    img_paths = glob.glob(os.path.join(prefix , "*.jpg"))

    np.random.shuffle(img_paths)
    for img_path in img_paths:
        img = image.load_img(img_path, target_size=(256, 256))
        x = image.img_to_array(img)
        inner_path = img_path.split('/')[1]
        label1 = int(inner_path.split('_')[0])
        label2 = int(inner_path.split('_')[1])
        images.append(x)
        labels['age_label'].append(label1)
        labels['gender_label'].append(label2)
    
    X = np.array(images, dtype= 'float32')
    Y_df = pd.DataFrame(labels)
    Y = Y_df.values
    return X, Y
```

```python
X_original, Y_original = readData('UTKFace/')

```

```python


def convert_age(age):
    if(age >= 1 and age <= 5): 
        return str(1)
    elif(age >= 6 and age <= 10): 
        return str(2)
    elif(age >= 11 and age <= 15): 
        return str(3)
    elif(age >= 16 and age <= 20): 
        return str(4)
    elif(age >= 21 and age <= 25): 
        return str(5)
    elif(age >= 26 and age <= 30): 
        return str(6)
    elif(age >= 31 and age <= 35): 
        return str(7)
    elif(age >= 36 and age <= 40): 
        return str(8)
    elif(age >= 41 and age <= 45): 
        return str(9)
    elif(age >= 46 and age <= 50): 
        return str(10)
    elif(age >= 51 and age <= 55): 
        return str(11)
    elif(age >= 56 and age <= 60): 
        return str(12)
    elif(age >= 61 and age <= 65): 
        return str(13)
    elif(age >= 66 and age <= 70): 
        return str(14)
    elif(age >= 71 and age <= 75): 
        return str(15)
    elif(age >= 76 and age <= 80): 
        return str(16)
    elif(age >= 81 and age <= 85): 
        return str(17)
    elif(age >= 86 and age <= 90): 
        return str(18)
    elif(age >= 91 and age <= 95): 
        return str(19)
    elif(age >= 96 and age <= 100): 
        return str(20)
    elif(age >= 101 and age <= 105): 
        return str(21)
    elif(age >= 106 and age <= 110): 
        return str(22)
    elif(age >= 111 and age <= 116): 
        return str(23)
    else: 
        return 'OOR'

```

```python
Y_original_df['age_group_label'] = Y_original_df['age_label'].map(convert_age)
```

```python
import h5py
from pathlib import Path
```

```python
hdf5_dir = Path("data/hdf5/")
hdf5_dir.mkdir(parents=True, exist_ok=True)
```

```python
def store_many_hdf5(images, labels):
    
    num_images = len(images)

    # Create a new HDF5 file
    file = h5py.File(hdf5_dir / f"{num_images}_many.h5", "w")

    # Create a dataset in the file
    dataset = file.create_dataset(
        "images", np.shape(images), h5py.h5t.STD_U8BE, data=images
    )
    meta_set = file.create_dataset(
        "meta", np.shape(labels), h5py.h5t.STD_U8BE, data=labels
    )
    file.close()
```
