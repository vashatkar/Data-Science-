### Defined functions for metrics

```python
def metrics(model, X_test, Y_test):
    snn_pred = model.predict(X_test, batch_size=32, verbose=1) 
    snn_predicted = np.argmax(snn_pred, axis=1)
    
    model_metrics= model.evaluate(X_test, Y_test)
    print(model)
    print("model accuracy on test data is : {}" .format(model_metrics[1]))
    #Confusion Matrix
    snn_cm = confusion_matrix(np.argmax(Y_test, axis=1), snn_predicted) 

    # Visualization
    snn_df_cm = pd.DataFrame(snn_cm, range(8), range(8)) 
    plt.figure(figsize = (20,14)) 
    sn.set(font_scale=1.4) #for label size 
    sn.heatmap(snn_df_cm, annot=True, annot_kws={"size": 12}) # font size 
    plt.show()
    
    snn_report = classification_report(np.argmax(Y_test, axis=1), snn_predicted)
    print(snn_report)
    
    # Plot linewidth.
    lw = 2

    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(num_classes):
        fpr[i], tpr[i], _ = roc_curve(Y_test[:, i], snn_pred[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Compute micro-average ROC curve and ROC area
    fpr["micro"], tpr["micro"], _ = roc_curve(Y_test.ravel(), snn_pred.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    # Compute macro-average ROC curve and ROC area

    # First aggregate all false positive rates
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(num_classes)]))

    # Then interpolate all ROC curves at this points
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(num_classes):
        mean_tpr += interp(all_fpr, fpr[i], tpr[i])

    # Finally average it and compute AUC
    mean_tpr /= num_classes

    fpr["macro"] = all_fpr
    tpr["macro"] = mean_tpr
    roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

    # Plot all ROC curves
    plt.figure(1, figsize=(15, 10), dpi=80)
   
    colors = cycle(['aqua', 'darkorange', 'cornflowerblue', 'pink', 'black', 'red', 'yellow', 'green'])
    for i, color in zip(range(num_classes), colors):
        plt.plot(fpr[i], tpr[i], color=color, lw=lw,
                label='ROC curve of class {0} (area = {1:0.2f})'
                ''.format(i, roc_auc[i])) 

    plt.plot([0, 1], [0, 1], 'k--', lw=lw)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic to multi-class')
    plt.legend(loc="lower right")
    plt.show()
```

```python
def convert_age3(age):
    if(age >= 1 and age <= 5):
        return str(1)
    elif(age >= 6 and age <= 18):
        return str(2)
    elif(age >= 19 and age <= 25):
        return str(3)
    elif(age >= 26 and age <= 35):
        return str(4)
    elif(age >= 36 and age <= 45):
        return str(5)
    elif(age >= 46 and age <= 50):
        return str(6)
    elif(age >= 51 and age <= 65):
        return str(7)
    elif(age >= 66 and age <= 116):
        return str(8)
    else:
        return 'OOR'
```

```python
# This will do preprocessing and realtime data augmentation:
datagen = ImageDataGenerator(
            # set input mean to 0 over the dataset
            featurewise_center=True,              ### changed to true
            # set each sample mean to 0
            samplewise_center=False,
            # divide inputs by std of dataset
            featurewise_std_normalization=True,        ### changed to true
            # divide each input by its std
            samplewise_std_normalization=False,
            # apply ZCA whitening
            zca_whitening=False,
            # epsilon for ZCA whitening
            zca_epsilon=1e-06,
            # randomly rotate images in the range (deg 0 to 180)
            rotation_range=10,
            # randomly shift images horizontally
            width_shift_range=0.1,
            # randomly shift images vertically
            height_shift_range=0.1,
            # set range for random shear
            shear_range=0.2,
            # set range for random zoom
            zoom_range=0,
            # set range for random channel shifts
            channel_shift_range=0.,
            # set mode for filling points outside the input boundaries
            fill_mode='nearest',
            # value used for fill_mode = "constant"
            cval=0.,
            # randomly flip images
            horizontal_flip=True,
            # randomly flip images
            vertical_flip=False,
            # set rescaling factor (applied before any other transformation)
            rescale=None,
            # set function that will be applied on each input
            preprocessing_function=None,
            # image data format, either "channels_first" or "channels_last"
            data_format='channels_last',
            # fraction of images reserved for validation (strictly between 0 and 1)
            validation_split=0.2)

        
```

```python
# Run training, with or without data augmentation.

def train_func(model ,x_train, y_train, datagen , data_augmentation = False, batch_size = 64, epochs = 10, validation_split =0.2 , \
               callbacks = None,\
               root_path='' , save_model = True, model_name=''  ):
    if not data_augmentation:
        print('Not using data augmentation.')
        model_history = model.fit(x_train, y_train,
              batch_size=batch_size,
              epochs=epochs,
              validation_split=validation_split,
              shuffle=True,
              callbacks=callbacks)
        data_aug = 'noaug'
    else:
        print('Using real-time data augmentation.')
        
        print("fitting the model using fit_generator !!!!")
        # Fit the model on the batches generated by datagen.flow().
        model_history = model.fit_generator(datagen.flow(x_train, y_train, batch_size=batch_size, subset='training'),
                        validation_data = datagen.flow(x_train, y_train, batch_size=batch_size, subset='validation') ,
                        epochs=epochs, shuffle=True,workers = 4, 
                        callbacks=callbacks)
        data_aug = 'aug'
    if save_model:
        model.save(root_path + 'saved_models/' + "{}_{}_{}_{}.h5".format(model_name,data_aug, epochs, batch_size ))    
    return model_history
```

```python
def tf_crop_and_resize_batches_of_arrays_mine(image_input, boxes_input , crop_size ):
    box_ind_processed = tf.range(tf.shape(image_input)[0])
    
    # print("shape of box_ind {}" .format(box_ind_processed))
    # batch_ids = tf.squeeze(tf.slice(boxes_input, [0, 0], [-1, 1], name="batch_id"))
    # num_boxes = tf.shape(boxes_input)[0]
    # batch_inds = tf.zeros([num_boxes], dtype=tf.int32, name='batch_inds')
    
    cropped_image = tf.image.crop_and_resize(image_input,
                        boxes_input, box_indices = box_ind_processed,
                        crop_size = (8,8) , method='bilinear' )
    return cropped_image
    
```

```python
squeezor2 = Lambda(lambda x: K.squeeze(x,axis=-1), name = 'squeezor2')
expandor = Lambda(lambda x: K.expand_dims(x, axis=-1) , name = 'expandor')
concatenator = Lambda(lambda x: K.concatenate(x, axis=-1), name = 'concatenator')
# lstm_cell = LSTM(128, activation='tanh', recurrent_activation='sigmoid',   \
#                                              return_sequences=False, return_state=True, stateful=False, unroll=False)


lstm_cell = LSTM(128, return_state=True)




# one_step_attention_layer = Lambda(one_step_attention, name='attention_layer')
```

```python
# normalizer = Lambda(lambda x: x/float(tf.reduce_sum(x)), name = 'normalizer')
```

```python
def keras_crop_and_resize_batches_of_arrays(image_input, boxes_input, crop_size=(10, 10)):
    def f_crop(packed):
        image, boxes = packed
        return tf_crop_and_resize_batches_of_arrays_mine(image, boxes, crop_size)
    return Lambda(f_crop , name = 'crop_and_resize')([image_input, boxes_input])
```

```python
# state_h = np.zeros((18966, 128))
# state_c = np.zeros((18966, 128))
# lstm_output = np.zeros((18966, 128))
```

```python
def alresnet_model_imp(input_shape ):

    # resnet50_input_tensor = Input(input_shape, name = 'resnet50_input_tensor')
    
    # state_h_0 = Input((128,), name = 'state_h_0')
    # state_c_0 = Input((128,), name = 'state_c_0')
    # lstm_output_0 = Input((128,), name = 'lstm_output_0')
    # state_h = state_h_0
    # state_c = state_c_0
    # lstm_output = lstm_output_0
    
    resnet50_input = resnet50_model.layers[0].get_layer(index = 0).input
    resnet50_output = resnet50_model.layers[0].get_layer(index = -2).output
    
    resnet50_inside = Model(inputs=resnet50_input, outputs=resnet50_output)
    
    for resnet_layer in resnet50_inside.layers:
        resnet_layer.trainable = False

    X = resnet50_inside(resnet50_input)
    
    X1 = GlobalAveragePooling2D()(X)
    
    X1 = Dense(512, activation='relu' , name = 'fully_connected_global')(X1)
    
    
    X1 = Reshape((512, 1))(X1)
    
    # context, attn_states = attn_layer([X1, lstm_output])
    
    # context = one_step_attention(X1, lstm_output)
    
    lstm_output, state_h , state_c = lstm_cell(X1)
    
    # lstm_output, state_h , state_c  = lstm_cell(context, initial_state = [state_h, state_c])
    
    state_hc = concatenator([state_h, state_c])
    state_hc = expandor(state_hc)

    L_4 = Conv1D(filters = 1, kernel_size = 250 , strides = 2, padding='valid',activation = 'sigmoid')(state_hc)
    

    L_4_squeezed = squeezor2(L_4)

    # print(L_4_squeezed.shape)


    # cropped_image = tf_crop_and_resize_batches_of_arrays_mine(X, L_4_squeezed, (8, 8))

    cropped_image = keras_crop_and_resize_batches_of_arrays(X, L_4_squeezed, crop_size=(8, 8))
    
    # print(cropped_image.shape)
    
    X_final = AveragePooling2D(pool_size=(8, 8))(cropped_image)
    
    X_final = Flatten()(X_final)
    
    X_final = Dense(num_classes, activation='softmax')(X_final)
    
    # print(X_final.shape)
    
    alresnet_model = Model(inputs = resnet50_input, outputs = X_final)
    
    return alresnet_model
    
```

## AL-Resnet based on both Global and Local Features

```python
# Prepare model model saving directory.

model_type = '8class%d' % (version)

model_name = 'big_model8_%s_model.{epoch:03d}.h5' % model_type
filepath = os.path.join(save_dir, model_name)

# Prepare callbacks for model saving and for learning rate adjustment.
big_model8_checkpoint = ModelCheckpoint(filepath=filepath,
                             monitor='val_accuracy',
                             verbose=1,
                             save_best_only=True)
big_model8_callbacks = [big_model_checkpoint]
```

```python
W1 = Lambda(lambda x: x , name = 'W1_mul')
W2_mul = Lambda(lambda x: x * (0.5) , name = 'W2_mul')
W_div = Lambda(lambda x: x/float(1.5) , name = 'W_div')
```

```python
def big_model(input_shape, alresnet_model):
    
    resnet_input = Input(input_shape, name ='resnet_input')
    
    # resnet50_input = resnet50_model.get_layer(index=-2).output
    # resnet50_output = resnet50_model.get_layer(index=-2).output
    
    resnet50_input = resnet50_model.layers[0].get_layer(index = 0).input
    resnet50_output = resnet50_model.layers[0].get_layer(index = -2).output
    
    resnet50_global = Model(inputs=resnet50_input, outputs=resnet50_output, name = 'AL-Resnet_Global')
    
    for resnet_layer in resnet50_global.layers:
        resnet_layer.trainable = False
    
    resnet50_output = resnet50_global(resnet_input)
    
    X= resnet50_output
    
    X1 = GlobalAveragePooling2D()(X)
    # print(X.shape)
    
    # X = Flatten()(X)
    
    X1 = Dense(256, activation = 'relu', name = 'fully_connected_X1')(X1)
    
    X1 = Dense(num_classes, activation = 'softmax' , name = 'softmax_X1')(X1)
    
    
    alresnet_input = alresnet_model.layers[0].input
    alresnet_output = alresnet_model.layers[-2].output
    alresnet_local_model = Model(inputs = alresnet_input ,
                                 outputs = alresnet_output, name = 'AL-Resnet-Local')
    
    
    
    alresnet_X = alresnet_local_model(resnet_input)
    
    #alresnet_X = alresnet_model.layers[-3].output
    
    X2 = Dense(256, activation = 'relu',  name = 'fully_connected_X2')(alresnet_X)
    
    X2 = Dense(num_classes, activation = 'softmax' , name = 'softmax_X2')(X2)
    
    X2_mul = W2_mul(X2)
    
    X_final = Add()([X1 , X2_mul])
    
    X_final = W_div(X_final)
    
    big_model = Model(inputs = resnet_input , outputs = X_final, name = 'AL-Resnet-Full')
    
    return big_model
    
```
